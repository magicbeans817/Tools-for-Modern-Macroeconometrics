```r
#if  (!require(forecast)) install.packages("forecast")
#if  (!require(forecast)) install.packages("uroot")


#library(stats)
#library(repr) 
#library(tseries)
#library(readr)
#library(forecast)
#library(uroot)

AR1.005.st <- 0
AR1.001.st <- 0
AR1.010.st <- 0

AR1.005.nst <- 0
AR1.001.nst <- 0
AR1.010.nst <- 0

AR2.005.st <- 0
AR2.001.st <- 0
AR2.010.st <- 0

AR2.005.nst <- 0
AR2.001.nst <- 0
AR2.010.nst <- 0

AR3.005.st <- 0
AR3.001.st <- 0
AR3.010.st <- 0

AR3.005.nst <- 0
AR3.001.nst <- 0
AR3.010.nst <- 0

p.valAR1 <- vector()
p.valAR2 <- vector()
p.valAR3 <- vector()

rep <- 2000 # of repetitions

for(seed in 1 : rep) {  
    
    l <- 100 
    e <- rnorm(l)
    
    #1st AR(1) process - w/ unit root
    set.seed(seed)                                                  #setting seed for replicability                                                     
    yt1 <- vector()                                                  #create blanc vector to store values of yt for t=1,...,l
    phi11 <- 1                                                       #set value of coefficient phi 1                                                    
    yt1[1] <- 0                                                      #set value of yt for t=1
    yt1[2] <- phi11 * e[1]                                            #set value of yt for t=2
    
    for (i in 3 : l){                                               #loop to fill in yt data for 2<t<l
        yt1[i] <- phi11 * yt1[i - 1] + e[i]
    }                                       
    
    pval1 <- adf.test(yt1)$p.value
    p.valAR1[seed] <- pval1
    
    if (pval1 <= 0.01) {
       AR1.001.st <- AR1.001.st + 1    
    }
    if (pval1 > 0.01) {
       AR1.001.nst <- AR1.001.nst + 1    
    }
    
    if (pval1 <= 0.05) {
       AR1.005.st <- AR1.005.st + 1    
    }
    if  (pval1 > 0.05) {
       AR1.005.nst <- AR1.005.nst + 1    
    }    
        
    if (pval1 <= 0.1) {
       AR1.010.st <- AR1.010.st + 1    
    }
    if (pval1 > 0.1){
       AR1.010.nst <- AR1.010.nst + 1    
    }

    #2nd AR(1) process - close to unit root
    set.seed(seed)                                                  #setting seed for replicability                                                      
    yt2 <- vector()                                                  #create blanc vector to store values of yt for t=1,...,l
    phi12 <- 0.95                                                      #set value of coefficient phi 1                                                    
    yt2[1] <- 0                                                      #set value of yt for t=1
    yt2[2] <- phi12 * e[1]                                            #set value of yt for t=2
    
    for (i in 3 : l){                                               #loop to fill in yt data for 2<t<l
        yt2[i] <- phi12 * yt2[i - 1] + e[i]
    }
    
    pval2 <- adf.test(yt2)$p.value
    p.valAR2[seed] <- pval2

    if (pval2 <= 0.05){
       AR2.005.st <- AR2.005.st + 1    
    }
    if (pval2 > 0.05){
       AR2.005.nst <- AR2.005.nst + 1    
    }
        
    if (pval2 <= 0.01){
       AR2.001.st <- AR2.001.st + 1    
    }
    if (pval2 > 0.01){
       AR2.001.nst <- AR2.001.nst + 1    
    }
           
    if (pval2 <= 0.1){
       AR2.010.st <- AR2.010.st + 1    
    }
    if (pval2 > 0.1){
       AR2.010.nst <- AR2.010.nst + 1    
    }

    #3rd AR(1) process - close to unit root
    set.seed(seed)                                                  #setting seed for replicability
    yt3 <- vector()                                                  #create blanc vector to store values of yt for t=1,...,l
    phi13 <- 0.9                                                       #set value of coefficient phi 1                                                    
    yt3[1] <- 0                                                      #set value of yt for t=1
    yt3[2] <- phi13 * e[1]                                            #set value of yt for t=2
    
    for (i in 3 : l){                                               #loop to fill in yt data for 2<t<l
        yt3[i] <- phi13 * yt3[i - 1] + e[i]
    }
    
    pval3 <- adf.test(yt3)$p.value
    p.valAR3[seed] <- pval3
    
    if (pval3 <= 0.05){
       AR3.005.st <- AR3.005.st + 1    
    }
    if (pval3 > 0.05){
       AR3.005.nst <- AR3.005.nst + 1    
    }
      
    if (pval3 <= 0.01){
       AR3.001.st <- AR3.001.st + 1    
    }
    if (pval3 > 0.01){
       AR3.001.nst <- AR3.001.nst + 1    
    }
           
    if (pval3 <= 0.1){
       AR3.010.st <- AR3.010.st + 1    
    }
    if (pval3 > 0.1){
       AR3.010.nst <- AR3.010.nst + 1    
    }
}


WRONG_DECISIONS <- matrix(c((AR1.001.nst+AR2.001.st+AR3.001.st)/(3*rep), c(AR1.001.st+AR2.001.nst+AR3.001.nst)/(3*rep), AR1.001.nst/rep, AR1.001.st/rep, AR2.001.st/rep, AR2.001.nst/rep, AR3.001.st/rep, AR3.001.nst/rep, 
                            (AR1.005.nst+AR2.005.st+AR3.005.st)/(3*rep), c(AR1.005.st+AR2.005.nst+AR3.005.nst)/(3*rep), AR1.005.nst/rep, AR1.005.st/rep, AR2.005.st/rep, AR2.005.nst/rep, AR3.005.st/rep, AR3.005.nst/rep, 
                            (AR1.010.nst+AR2.010.st+AR3.010.st)/(3*rep), c(AR1.010.st+AR2.010.nst+AR3.010.nst)/(3*rep), AR1.010.nst/rep, AR1.010.st/rep, AR2.010.st/rep, AR2.010.nst/rep, AR3.010.st/rep, AR3.010.nst/rep),ncol=12)
colnames(WRONG_DECISIONS) <- c("0.01","AR1","AR2","AR3","0.05","AR1","AR2","AR3","0.1","AR1","AR2","AR3")
rownames(WRONG_DECISIONS) <- c("Correct","Wrong")
WRONG_DECISIONS <- as.table(WRONG_DECISIONS)
WRONG_DECISIONS

DT <- matrix(c(AR1.001.st, AR1.001.nst, AR1.001.nst+AR1.001.st, AR1.005.st, AR1.005.nst, AR1.005.nst+AR1.005.st, AR1.010.st, AR1.010.nst, AR1.010.nst+AR1.010.st, AR2.001.st, AR2.001.nst, AR2.001.nst+AR2.001.st, AR2.005.st, AR2.005.nst, AR2.005.nst+AR2.005.st, AR2.010.st, AR2.010.nst, AR2.010.nst+AR2.010.st, AR3.001.st, AR3.001.nst, AR3.001.nst+AR3.001.st, AR3.005.st, AR3.005.nst, AR3.005.nst+AR3.005.st, AR3.010.st, AR3.010.nst, AR3.010.nst+AR3.010.st), nrow=3)
rownames(DT) <- c("stationary","nonstationary","total")
colnames(DT) <- c("AR1-0.01","AR1-0.05","AR1-0.10","AR2-0.01","AR2-0.05","AR2-0.10","AR3-0.01","AR3-0.05","AR3-0.10")
DT <- as.table(DT)
DT

P.VAL <- matrix(c(p.valAR1,p.valAR2,p.valAR3),ncol=3)
colnames(P.VAL) <- c("AR1","AR2","AR3")
rownames(P.VAL) <- c(1:rep)
P.VAL <- as.table(P.VAL)
#P.VAL
```
