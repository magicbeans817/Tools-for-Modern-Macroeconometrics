## 1) On an empirical time series (of your choice) perform the Perron' test for structural change (see Enders 2014, p. 229). Comment the results.

As my empirical data, I employ python script which downloads google trends data for keyowrd "covid" for the last three months. That may be interesting, especially considering introduction of omicron variant and invasion of Ukraine in last days. Let's take a look at the data:

```r
data <- read.csv("search_trends.csv")
data <- ts(data$covid)

data %>% glimpse()
print(paste("Number of missing observations is :", sum(is.na(data))))

plot.ts(data, ylab = NA)
length(data)

par(mfrow = c(1, 2))
acf(data)
pacf(data)
```

According to what I see from ACF, PACF and simple visualization, it seems that the data might be non-stationary, but I need more evidence to claim that. Thus I will employ the Perron' test for structural change to arrive to conclusions.

```r
tseries::pp.test(data, alternative = c("stationary"), lshort = FALSE)
```

I am not able to reject the null hypothesis. Therefore I assume non-stationarity of my data. I can try to difference the data and perform the same test again:

```r
data_d <- diff(data)

plot.ts(data_d, ylab = NA)

par(mfrow = c(1, 2))
acf(data_d)
pacf(data_d)
```

```r
tseries::pp.test(data_d, alternative = c("stationary"), lshort = TRUE)
```

As can be seen above, first differencing highly likely solved the non-stationarity problem of the data, hence from now I have enough evidence to consider the data as stationary.

## 2) On a time series (real or simulated) estimate breaks (structural change). Discuss an importance of chosen model in the estimation routines. Comment the results.

I have chosen to simulate the data, as it might be interesting to see drawbacks of the methods in contrast with knowing how the data were obtained. I will simulate the data in such a way that there will be a break at half of the data.

```r
set.seed(69420)                                                        # first set seed for replicability
n <- 100                                                             # number of observations
e <- rnorm(n)                                                          # error terms
y <- vector()                                                          # blank vector to store values of explained variable


phi_1 <- 0.70                                                         # set value of parameter
mu_1 <- 2

phi_2 <- 0.5
mu_2 <- 5

y[1] <- 0                                                            # set value of explained variable in first period

for (i in 2:(0.5 *n)){
    y[i] <- mu_1 + phi_1 * y[i-1] + e[i]                                      # create the rest of eplained variable
}

for (i in (n/2 + 1):n){
    y[i] <- mu_2 + phi_2 * y[i - 1] + e[i]
}

options(repr.plot.width = 10, repr.plot.height = 4)
plot.ts(y, ylab = NA)   
```

Plot and shape of the curve imply possible break exactly at half of the data, as was my intention. Now I will plot ACF, PACF, estimate the model and plot ACF and PACF for estimated residuals.

```r
par(mfrow = c(1, 2))
acf(y)
pacf(y)

model <- Arima(y, order = c(1,0,0))
summary(model)

par(mfrow = c(1, 2))
acf(model$residuals, main = NA)
pacf(model$residuals, main = NA)
```

Both coefficients of the model appear to be significant, however their estimated values correspond to neither first half value nor second half values. Lets forecast and see how the good performs when forecasting:

```r
plot.ts(y, ylab = NA, main = 'Fitted values')
lines(model$fitted, col = 'red')
```

It seems  that the model is not useless, nevertheless it is obvious that something does not work very well. Now I will estimate the structural break. I will begin by applying Chow test. With simulated data, it makes sense to apply Chow test, as I am sure where in the data the structural breaks is.

```r
trend<-c(1:n)
sctest(y ~ trend, type = "Chow", point = (n/2))
```

Chow test supports alternative hypothesis as expected. Next, I will apply Zivot and Andrews Unit Root Test.

```r
z_a <-ur.za(y, model = c("intercept", "trend", "both"), lag= 1)
summary(z_a)
plot(z_a)
```

Zivot and Andrews test correctly estimates the breakpoint. Now I will use breakpoints function. The maximum bumber of possible breaks to be estimated is set to one, as there is only one break.

```r
brk <- breakpoints(ts(y[2:n]) ~ y[1:(n-1)], h = 20, breaks = 1)

summary(brk)
```

Breakpoints function does not disappoint and returns back what is expeted to be returned. Considering results of the test, I will now estimate 2 models, one for each half of the data. Last, I will make forecasts.

```r
model_1half <- Arima(y[1:(n/2)], order = c(1,0,0))
summary(model_1half)

model_2half <- Arima(y[(n/2 + 1):n], order = c(1,0,0))
summary(model_2half)

plot.ts(y, ylab = NA, main = 'Fitted values')
lines(c(model_1half$fitted, model_2half$fitted) , col = 'red')
```

In comparison with estimating whole dataset, estimating parts of the dataset individually before and after the breakpoint seems to make more sense. Also, both models achieve much lower AICc.
