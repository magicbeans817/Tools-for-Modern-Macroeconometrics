### 2. Find turning points of your time series. Use sequential application of logical rules mentioned on slide 10. What's the average duration and recession? Compare it with the Hodrick-Prescott filter.

Lets difference the time series  in order to get rid of trend.

```r
diff_log <- diff(log_rgdp) * 100

options(repr.plot.width = 10, repr.plot.height = 6)

plot.ts(diff_log , lwd = 0.1)
# lty = "dashed", lwd = 3,
abline(h = 0, col = "darkgreen")
```

I will start with A. Okun (1962) to identify 2 or more subsequent quarters of negative gdp growth. Considering length of the series, I will rather observe and compute it manually than spend two hours figuring out how to automate the process.

```r
tf <- diff_log < 0
tf
```

```r
delka_recesi <- c(7, 3, 2, 3, 5, 2, 2, 2, 4)
mean(delka_recesi)
```

Thus the mean length of recession according to Okun's approach would be three and one third of a quarter. In the data, there are three recessions of length longer than is the average. First of these three recessions is connected to beginning of 1990s (probably caused by split of Soviet Union), second to 2008 crisis and the last one to Covid-19 crisis. Otherwise lengths of recessions never surpassed more than three quarters. This would imply that finnish economic is rather capable of of dealing with turmoils caused by business cycle and only strong external stochastic shocks from the outside cause more significant trouble.

Now I will move to identifying potential peaks and throughs.

```r
# Pro prehlednost a konzistenci si vsechna data hodim do jedne tabulky
data <- finland
data$log_rgdp <- log_rgdp
data$diff_log <- c(NA, diff_log)
data %>% head
```

```r
data$peaks <- 0

for (i in 3:(nrow(data)-2)){
    x <- c(data[i - 2, "real_gdp"], data[i - 1, "real_gdp"], data[i, "real_gdp"],
           data[i + 1, "real_gdp"], data[i + 2, "real_gdp"])
    if(x[3] == max(x)){
        data[i, "peaks"] <- 1
    }
}

data$trough <- 0

for (i in 3:(nrow(data)-2)){
    x <- c(data[i - 2, "real_gdp"], data[i - 1, "real_gdp"], data[i, "real_gdp"],
           data[i + 1, "real_gdp"], data[i + 2, "real_gdp"])
    if(x[3] == min(x)){
        data[i, "trough"] <- - 1
    }
}

data$alternate <- data$peaks + data$trough

data_sub <- subset(data, alternate != 0)
data_sub
```

This shows that peaks and troughs alter in my whole dataset with one exception. This is actually nice result, I will plot it to make it more clear:

```r
plot.ts(data$real_gdp, ylab = "")
title("Real GDP")

peaks <- subset(data, peaks == 1)
points(peaks$date, peaks$real_gdp, pch = 1, col = "red", cex = 2)

troughs <- subset(data, trough == -1)
points(troughs$date, troughs$real_gdp, pch = 1, col = "blue", cex = 2)

legend("topleft", c("Peaks", "Troughs"),
       col=c("red", "blue"), lwd=3)
```

### 3. Get another time series with higher frequency, such as industrial production, unemployment. Estimate a model that will mix both frequencies. It's up to you whether to opt for the MIDAS model or the structural time series model. Grab fitted values for 2021Q4 and make a prediction for 2022Q1.

```r
unemp <- read.csv("DP_LIVE_28032022211205262.csv")
unemp <- unemp[,c("TIME", "Value")] # subset only the important stuff

gdp <- read.csv("CLVMNACSCAB1GQFI.csv")
colnames(gdp) <- c("date", "real_gdp")

gdp <- ts(gdp$real_gdp, frequency = 4, start = c(1990, 1))
unemp <- ts(unemp$Value, frequency = 12, start = c(1990, 1))
```

```r
gdp_x <- window(gdp, end = c(2021, 4))
unemp_y <- window(unemp, end = c(2021, 12))

gdp_ld <- diff(log(gdp_x))*100
unemp_ld <- diff(log(unemp_y)) * 10

# Graph of the data
plot.ts(gdp_ld, xlab = "Time", ylab = "Percentages", col = "blue", ylim = c(-5, 6))
lines(unemp_ld, col = "red")
```

```r
beta0 <- midasr::midas_r(gdp_ld ~ mls(gdp_ld, 1, 1) + mls(unemp_ld, 3:11, 3, nbeta), start = list(unemp_ld = c(1.7, 1, 5)))
coef(beta0) #prints the estimated coefficients
```
